```python
import pandas as pd
```

```python
df = pd.read_csv("artist.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[["artist_name","followers"]]
```

```python
df
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
top_10 = df.nlargest(10,"followers")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_10['artist_name'], top_10['artist_name'], color='blue')
plt.xlabel('artist_name')
plt.ylabel('followers')
plt.title('Top 10 Artists with Highest Followers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
